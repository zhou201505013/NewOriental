```python
# 每日情感值，不同平台情感值图
%matplotlib inline
```

```python
import pandas as pd

sentiment = pd.read_csv('../data/sentiment_value.csv')
source_grouped = sentiment.groupby(['source'])
day_grouped = sentiment.groupby(['date'])
```

```python
data_source = [source_grouped.get_group('bilibili')['sentiment_value'],
               source_grouped.get_group('toutiao')['sentiment_value'],
               source_grouped.get_group('weibo')['sentiment_value'],
               source_grouped.get_group('douyin')['sentiment_value']]
len(data_source), len(data_source[0])
```

```python
data_day = [day_grouped.get_group('2022-06-07')['sentiment_value'],
            day_grouped.get_group('2022-06-08')['sentiment_value'],
            day_grouped.get_group('2022-06-09')['sentiment_value'],
            day_grouped.get_group('2022-06-10')['sentiment_value'],
            day_grouped.get_group('2022-06-11')['sentiment_value'],
            day_grouped.get_group('2022-06-12')['sentiment_value'],
            day_grouped.get_group('2022-06-13')['sentiment_value'],
            day_grouped.get_group('2022-06-14')['sentiment_value'],
            day_grouped.get_group('2022-06-15')['sentiment_value'],
            day_grouped.get_group('2022-06-16')['sentiment_value'],
            day_grouped.get_group('2022-06-17')['sentiment_value'],
            day_grouped.get_group('2022-06-18')['sentiment_value'],
            day_grouped.get_group('2022-06-19')['sentiment_value'],
            day_grouped.get_group('2022-06-20')['sentiment_value']
            ]
len(data_day), len(data_day[0])
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

fig1, ax1 = plt.subplots(figsize=(17, 4))


# plot violin plot
ax1.violinplot(data_day,
                  showmeans=False,
                  showmedians=True)
#ax1.set_title('Daily sentiment value')
ax1.yaxis.grid(True)
ax1.set_xticks([y + 1 for y in range(len(data_day))],
                  labels=['06-07', '06-08', '06-09', '06-10', '06-11', '06-12', '06-13',
                          '06-14', '06-15', '06-16', '06-17', '06-18', '06-19', '06-20'])
ax1.set_xlabel('日期')
ax1.set_ylabel('情感值')

plt.savefig('../images/每日情感值.svg', dpi=300)
plt.show()
```

```python
fig2, ax2 = plt.subplots(figsize=(4.5, 4))
ax2.boxplot(data_source)
#ax2.set_title('Sentiment value of different platforms')
ax2.yaxis.grid(True)
ax2.set_xticks([y + 1 for y in range(len(data_source))],
               labels=['哔哩哔哩', '今日头条', '新浪微博', '抖音'])
ax2.set_xlabel('平台')
ax2.set_ylabel('情感值')
plt.savefig('../images/不同平台情感值.svg', dpi=300)
plt.show()
```
