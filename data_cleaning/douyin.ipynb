```python
# 抖音数据处理
import pandas as pd
```

```python
douyin = pd.read_excel('data/抖音视频评论采集1_20220620100501164.xlsx')
douyin
```

```python
douyin = douyin.drop_duplicates(subset='评论内容')
douyin.shape
```

```python
douyin = douyin[~ douyin['评论内容'].isnull()]
douyin.shape
```

```python
douyin = douyin[~ douyin['评论内容'].str.contains('@')]
douyin.shape
```

```python
pd.Timestamp('2022.06.20')
```

```python
douyin['评论时间'] = douyin['评论时间'].apply(pd.Timestamp)
douyin
```

```python
douyin = douyin[douyin['评论时间'] > pd.Timestamp('2021-12-28')]
douyin
```

```python
douyin.to_excel('data/douyin_comment.xlsx')
data_douyin = pd.DataFrame()
data_douyin[['time', 'comment']] = douyin[['评论时间', '评论内容']]
data_douyin.to_excel('data/data_douyin.xlsx')
```
