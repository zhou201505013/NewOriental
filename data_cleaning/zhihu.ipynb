```python
# 知乎数据处理，未使用
import pandas as pd
```

```python
zhihu = pd.read_excel('data/temp_知乎回答评论_抓取_规则_1_列表_20220620002803159.xlsx')
zhihu
```

```python
zhihu = zhihu.drop_duplicates(subset='评论内容')
zhihu
```

```python
# 统一时间格式
# 06-17 昨天 12:23 22 小时前 2021-11-15
# 处理IP属地
zhihu_date = zhihu[zhihu['评论时间'].str.contains(r'IP\s')]
zhihu_yesterday = zhihu[zhihu['评论时间'].str.contains('昨天')]
zhihu_hour = zhihu[zhihu['评论时间'].str.contains('小时前')]
zhihu_date_year = zhihu[zhihu['评论时间'].str.contains(r'\d{4}-\d{2}-\d{2}')]
zhihu_date.shape, zhihu_yesterday.shape, zhihu_hour.shape, zhihu_date_year.shape
```

```python
pd.concat([zhihu, zhihu_date, zhihu_yesterday, zhihu_hour, zhihu_date_year]).drop_duplicates(subset='评论时间')
```
