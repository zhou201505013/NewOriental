```python
# LDA demo
import jieba
import pandas as pd
```

```python
data_bilibili = pd.read_excel('data/data_bilibili.xlsx')
data_bilibili
```

```python
bilibili_test_set = data_bilibili.sample(n=100)
bilibili_test_set
```

```python
toutiao = pd.read_excel('data/data_toutiao.xlsx')
toutiao
```

```python
toutiao_test_set = toutiao.sample(n=100)
toutiao_test_set
```

```python
bilibili_test_set['words'] = bilibili_test_set['comment'].apply(jieba.lcut)
bilibili_test_set
```

```python
toutiao_test_set['words'] = toutiao_test_set['comment'].apply(jieba.lcut)
toutiao_test_set
```

```python
bilibili_words = bilibili_test_set['words'].values
bilibili_words
```

```python
toutiao_words = toutiao_test_set['words'].values
toutiao_words
```

```python
with open('data/hit_stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.readlines()
stop_words
```

```python
stop_words = [stop_word.strip('\n') for stop_word in stop_words]
stop_words
```

```python
bilibili_test_set['sentence'] = bilibili_test_set['words'].apply(lambda sentence: list(filter(lambda word: word not in stop_words, sentence)))
bilibili_test_set
```

```python
toutiao_test_set['sentence'] = toutiao_test_set['words'].apply(lambda sentence: list(filter(lambda word: word not in stop_words, sentence)))
toutiao_test_set
```

```python
from gensim import corpora, models
```

```python
bilibili_dictionary = corpora.Dictionary(bilibili_test_set['sentence'].values)
toutiao_dictionary = corpora.Dictionary(toutiao_test_set['sentence'].values)
bilibili_dictionary
```

```python
print(bilibili_dictionary)
```

```python
bilibili_dictionary.token2id
```

```python
bilibili_corpus = [bilibili_dictionary.doc2bow(sentence) for sentence in bilibili_test_set['sentence']]
toutiao_corpus = [toutiao_dictionary.doc2bow(sentence) for sentence in toutiao_test_set['sentence']]
bilibili_corpus
```

```python
toutiao_corpus
```

```python
bilibili_lda = models.ldamodel.LdaModel(corpus=bilibili_corpus, num_topics=2, id2word=bilibili_dictionary, passes=20)
toutiao_lda = models.ldamodel.LdaModel(corpus=toutiao_corpus, num_topics=2, id2word=toutiao_dictionary, passes=20)
```

```python
bilibili_lda.print_topics(num_topics=2, num_words=3)
```

```python
toutiao_lda.print_topics(num_topics=2, num_words=3)
```
