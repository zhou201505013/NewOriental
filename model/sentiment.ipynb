```python
# SKEP 情感分析
import pandas as pd
import paddlehub as hub
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '0' 
```

```python
data = pd.read_csv('../data/data_all.csv')
module = hub.Module(name="ernie_skep_sentiment_analysis")
```

```python
data['sentiment_label'] = data['comment'].apply(lambda x: module.predict_sentiment([x], use_gpu=True)[0]['sentiment_label'])
data['positive_probs'] = data['comment'].apply(lambda x: module.predict_sentiment([x], use_gpu=True)[0]['positive_probs'])
data['negative_probs'] = data['comment'].apply(lambda x: module.predict_sentiment([x], use_gpu=True)[0]['negative_probs'])
data.head()
```

```python
data.sample(30)
```

```python
data.to_csv('../data/sentiment.csv')
```
